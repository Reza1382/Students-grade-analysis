```python
import numpy as np
import pandas as pd

df = pd.read_csv('student-mat.csv')
df['grade_mean'] = (df['G1'] + df['G2'] + df['G3']) / 3

# Analysis of Final Scores Mean
print(df["grade_mean"].describe())

# Filtering scores on based Sex
gender_stats = df.groupby('sex')['grade_mean'].agg(['mean', 'std', 'count'])
print(gender_stats)
print('='*60) 

# Analyse of Factors Affecting Student Grades
abcence_mean = df.groupby('absences')['grade_mean'].agg([ 'count', 'mean', 'std'])
print(abcence_mean)
print("correlation between absence numbers and student's mean of scores : ", 
      np.corrcoef(df["absences"], df['grade_mean'])[0, 1])                               # Effect of Number of Abcences
print('='*60) 

df['parents_edu'] = (df['Medu'] + df['Fedu']) / 2
parents_edu = df.groupby(["parents_edu"])['grade_mean'].agg([ 'count', 'mean', 'std'])
print(parents_edu)
print("correlation between parents education and student's mean of scores : ", 
      np.corrcoef(df["parents_edu"], df['grade_mean'])[0, 1])                             # Effect of Parents Education
print('='*60) 

school_sort = df.groupby(["school"])['grade_mean'].agg([ 'count', 'mean', 'std'])         # Effect of type of school
print(school_sort)                                                               
print('='*60)

mother_job = df.groupby(["Mjob"])['grade_mean'].agg([ 'count', 'mean', 'std'])           # Effect of Motehr's Job
print(mother_job)
print('='*60)

father_job = df.groupby(["Fjob"])['grade_mean'].agg([ 'count', 'mean', 'std'])           # Effect of Father's Job
print(father_job)
print('='*60) 

parents_job = df.groupby(["Fjob", 'Mjob'])['grade_mean'].agg([ 'count', 'mean', 'std'])  # Effect of Parents Job
print(parents_job)
print(parents_job['mean'].agg(['argmax', 'max', 'argmin', 'min'])) 
print('='*60)

triple_group = df.groupby(['sex', 'age'])['grade_mean'].agg(['mean', 'count'])          # group mean of scores by sex and age
print(triple_group.groupby('sex')['mean'].max())
```
